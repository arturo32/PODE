```python
import pandas as pd
import numpy as np
```

```python
estruturas_curriculares = pd.read_csv('dados_brutos/estruturas-curriculares.csv', sep=';')
curriculo_componente_graduacao = pd.read_csv('dados_brutos/curriculo-componente-graduacao.csv', sep=';')
componentes_curriculares_presenciais = pd.read_csv('dados_brutos/componentes-curriculares-presenciais.csv', sep=';')
```

```python
componentes_curriculares_presenciais.columns = ['id_componente_curricular', 'tipo_componente', 'codigo', 'nivel', 'nome',
       'unidade_responsavel', 'ch_teorico', 'ch_pratico', 'ch_estagio',
       'ch_total', 'ch_dedicada_docente', 'ch_ead', 'cr_max_ead',
       'permite_horario_flexivel', 'qtd_unidades', 'procedimentos_avaliacao',
       'equivalencia', 'pre_requisito', 'co_requisito', 'ementa',
       'bibliografia', 'objetivos', 'conteudo', 'competencias_habilidades',
       'referencias', 'ano_programa', 'periodo_programa', 'modalidade',
       'curso_componente']
```

```python
informacoes_estruturas_curriculares = [
    'id_curriculo', 'nome_matriz', 'nome_curso','semestre_conclusao_minimo', 'semestre_conclusao_ideal',
    'semestre_conclusao_maximo', 'ch_total_minima', 'ch_nao_atividade_obrigatoria', 'ch_optativas_minima','ch_complementar_minima',
    'max_eletivos','ch_minima_semestre','ch_maxima_semestre'
]
```

```python
cursos_ti = estruturas_curriculares[estruturas_curriculares['nome_curso'] == 'TECNOLOGIA DA INFORMAÇÃO'][informacoes_estruturas_curriculares]
cursos_ti
```

```python
cursos_ti.iloc[:, 2] = "BACHARELADO EM TECNOLOGIA DA INFORMAÇÃO"
cursos_ti.iloc[0, 1] = "NOTURNO"
cursos_ti.iloc[1, 1] = "DIURNO"
cursos_ti.iloc[2, 1] = "DESENVOLVIMENTO DE SOFTWARE"
cursos_ti.iloc[3, 1] = "COMPUTAÇÃO"
cursos_ti.columns = ['id_curriculo','nome_matriz','nome_curso','semestre_conclusao_minimo','semestre_conclusao_ideal','semestre_conclusao_maximo','ch_total_minima',
                     'ch_obrigatoria', 'ch_optativa_minima', 'ch_complementar_minima','ch_eletiva_maxima','ch_semestre_minima','ch_semestre_maxima']
```

```python
cursos_ti.iloc[:2,:].to_csv('dados_extraidos/cursos_ti.csv', index=False)
cursos_ti.iloc[2:,:].to_csv('dados_extraidos/enfases_ti.csv', index=False)
```

```python
curriculo_componente_ti = curriculo_componente_graduacao[curriculo_componente_graduacao.id_curriculo.isin(cursos_ti.id_curriculo)]
```

```python
curriculo_componente_ti_noturno = curriculo_componente_graduacao[curriculo_componente_graduacao.id_curriculo == 133797961]

curriculo_componente_ti_noturno_obrigatorias = curriculo_componente_ti_noturno[curriculo_componente_ti_noturno.tipo_vinculo_componente == "OBRIGATÓRIO"]
curriculo_componente_ti_noturno_obrigatorias.drop(['id_curriculo','id_curriculo_componente','tipo_vinculo_componente','nivel_ensino'], axis=1, inplace=True)
curriculo_componente_ti_noturno_obrigatorias = curriculo_componente_ti_noturno_obrigatorias.merge(componentes_curriculares_presenciais[['id_componente_curricular','codigo']]) 
curriculo_componente_ti_noturno_obrigatorias.drop(['id_componente_curricular'], axis=1, inplace=True)
curriculo_componente_ti_noturno_obrigatorias.to_csv("dados_extraidos/obrigatorias_noturno.csv", index=False)

curriculo_componente_ti_noturno_optativas = curriculo_componente_ti_noturno[curriculo_componente_ti_noturno.tipo_vinculo_componente == "OPTATIVO"]
curriculo_componente_ti_noturno_optativas.drop(['id_curriculo','id_curriculo_componente','semestre_oferta','tipo_vinculo_componente','nivel_ensino'], axis=1, inplace=True)
curriculo_componente_ti_noturno_optativas = curriculo_componente_ti_noturno_optativas.merge(componentes_curriculares_presenciais[['id_componente_curricular','codigo']]) 
curriculo_componente_ti_noturno_optativas.drop(['id_componente_curricular'], axis=1, inplace=True)
curriculo_componente_ti_noturno_optativas.to_csv("dados_extraidos/optativas_noturno.csv", index=False)
```

```python
curriculo_componente_ti_diurno = curriculo_componente_graduacao[curriculo_componente_graduacao.id_curriculo == 133795010]

curriculo_componente_ti_diurno_obrigatorias = curriculo_componente_ti_diurno[curriculo_componente_ti_diurno.tipo_vinculo_componente == "OBRIGATÓRIO"]
curriculo_componente_ti_diurno_obrigatorias.drop(['tipo_vinculo_componente','nivel_ensino'], axis=1, inplace=True)

curriculo_componente_ti_diurno_optativas = curriculo_componente_ti_diurno[curriculo_componente_ti_diurno.tipo_vinculo_componente == "OPTATIVO"]
curriculo_componente_ti_diurno_optativas.drop(['semestre_oferta','tipo_vinculo_componente','nivel_ensino'], axis=1, inplace=True)
```

```python
curriculo_componente_ti_dev_soft = curriculo_componente_graduacao[curriculo_componente_graduacao.id_curriculo == 134044403]

curriculo_componente_ti_dev_soft_obrigatorias = curriculo_componente_ti_dev_soft[curriculo_componente_ti_dev_soft.tipo_vinculo_componente == "OBRIGATÓRIO"]
curriculo_componente_ti_dev_soft_obrigatorias.drop(['tipo_vinculo_componente','nivel_ensino'], axis=1, inplace=True)

curriculo_componente_ti_dev_soft_optativas = curriculo_componente_ti_dev_soft[curriculo_componente_ti_dev_soft.tipo_vinculo_componente == "OPTATIVO"]
curriculo_componente_ti_dev_soft_optativas.drop(['semestre_oferta','tipo_vinculo_componente','nivel_ensino'], axis=1, inplace=True)
```

```python
curriculo_componente_ti_comput = curriculo_componente_graduacao[curriculo_componente_graduacao.id_curriculo == 133804382]

curriculo_componente_ti_comput_obrigatorias = curriculo_componente_ti_comput[curriculo_componente_ti_comput.tipo_vinculo_componente == "OBRIGATÓRIO"]
curriculo_componente_ti_comput_obrigatorias.drop(['tipo_vinculo_componente','nivel_ensino'], axis=1, inplace=True)

curriculo_componente_ti_comput_optativas = curriculo_componente_ti_comput[curriculo_componente_ti_comput.tipo_vinculo_componente == "OPTATIVO"]
curriculo_componente_ti_comput_optativas.drop(['semestre_oferta','tipo_vinculo_componente','nivel_ensino'], axis=1, inplace=True)
curriculo_componente_ti_comput_optativas.to_csv('dados_extraidos/curriculo_componente_ti_comput_optativas.csv', index=False)
```

```python
curriculo_componente_ti_obrigatorias = pd.concat([curriculo_componente_ti_noturno_obrigatorias, curriculo_componente_ti_diurno_obrigatorias, 
                                                  curriculo_componente_ti_dev_soft_obrigatorias, curriculo_componente_ti_comput_obrigatorias])
curriculo_componente_ti_obrigatorias = curriculo_componente_ti_obrigatorias.reset_index(drop=True)
curriculo_componente_ti_obrigatorias.to_csv('dados_extraidos/curriculo_componente_ti_obrigatorias.csv', index=False)

curriculo_componente_ti_optativas = pd.concat([curriculo_componente_ti_noturno_optativas, curriculo_componente_ti_diurno_optativas, 
                                                  curriculo_componente_ti_dev_soft_optativas, curriculo_componente_ti_comput_optativas])
curriculo_componente_ti_optativas = curriculo_componente_ti_optativas.reset_index(drop=True)
curriculo_componente_ti_optativas.to_csv('dados_extraidos/curriculo_componente_ti_optativas.csv', index=False)
```
