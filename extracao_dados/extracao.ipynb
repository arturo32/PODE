```python
import pandas as pd
```

```python
estruturas_curriculares = pd.read_csv('dados_brutos/estruturas-curriculares.csv', sep=';')
curriculo_componente_graduacao = pd.read_csv('dados_brutos/curriculo-componente-graduacao.csv', sep=';')
componentes_curriculares_presenciais = pd.read_csv('dados_brutos/componentes-curriculares-presenciais.csv', sep=';')
```

```python
informacoes_estruturas_curriculares = [
    'id_curriculo', 'nome_matriz', 'nome_curso','semestre_conclusao_minimo', 'semestre_conclusao_ideal',
    'semestre_conclusao_maximo', 'ch_total_minima', 'ch_nao_atividade_obrigatoria', 'ch_optativas_minima','ch_complementar_minima',
    'max_eletivos','ch_minima_semestre','ch_maxima_semestre'
]
```

```python
cursos_ti = estruturas_curriculares[estruturas_curriculares['nome_curso'] == 'TECNOLOGIA DA INFORMAÇÃO'][informacoes_estruturas_curriculares]
cursos_ti
```

```python
cursos_ti.to_csv('dados_extraidos/cursos_ti.csv')
```

```python
curriculo_componente_ti = curriculo_componente_graduacao[curriculo_componente_graduacao.id_curriculo.isin(cursos_ti.id_curriculo)]
```

```python
curriculo_componente_ti.to_csv("dados_extraidos/curriculo_componente_ti.csv")
```

```python
informacoes_componentes_curriculares_presenciais = [
    'id_componente', 'codigo', 'nome','ch_teorico', 'ch_pratico', 'ch_total',
    'equivalencia', 'pre_requisito','co_requisito'
]
```

```python
disciplinas_ti = componentes_curriculares_presenciais[informacoes_componentes_curriculares_presenciais]
```

```python
disciplinas_ti = disciplinas_ti[disciplinas_ti.id_componente.isin(curriculo_componente_ti.id_componente_curricular)]
```
